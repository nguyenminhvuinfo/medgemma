## Di chuyển

```python
import pandas as pd
import shutil
import os

drive_path = '/content/drive/MyDrive/thuc-tap/datasets/discharge.csv.gz'
local_dir = '/content/data'
local_path = os.path.join(local_dir, 'discharge.csv.gz')

if not os.path.exists(local_dir):
    os.makedirs(local_dir)

shutil.copy(drive_path, local_path)

df = pd.read_csv(local_path, compression='gzip')

print(df.head())
print(df.info())
```

## Lọc dữ liệu và lưu file

```python
import pandas as pd

patient_counts = df['subject_id'].value_counts()
valid_patients = patient_counts[patient_counts > 5].index
selected_patients = valid_patients[:100]

df_sample = df[df['subject_id'].isin(selected_patients)]

output_path = '/content/drive/MyDrive/thuc-tap/datasets/discharge_sample_100_patients_gt5.csv.gz'
df_sample.to_csv(output_path, index=False, compression='gzip')

print(f"Tổng số bệnh nhân gốc: {df['subject_id'].nunique()}")
print(f"Số bệnh nhân có trên 5 lần khám: {len(valid_patients)}")
print(f"Số bệnh nhân được chọn: {len(selected_patients)}")
print(f"Tổng số dòng ghi chú thu được: {len(df_sample)}")
print(f"File đã được lưu tại: {output_path}")
```
